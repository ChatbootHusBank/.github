# Add stopwords til den standard filen

```python
import spacy

nlp = spacy.load('nb_core_news_lg')

# Liste med nye stoppord
new_stopwords = ['housbank','det','man']

for word in new_stopwords:
    nlp.Defaults.stop_words.add(word) 
    nlp.vocab[word].is_stop = True 

#sjekke om de nye ordene er lagt til som stoppord
for word in new_stopwords:
    print(f"{word} is stop word: {word in nlp.Defaults.stop_words and nlp.vocab[word].is_stop}")
```

```python
import pandas as pd
import spacy

nlp = spacy.load("nb_core_news_lg")

df = pd.read_csv('Dataset_for_chatbot-2.csv', delimiter=';')

# Funksjon for å fjerne stopwords
def remove_stopwords(text):
    if pd.isnull(text):
        return text
    doc = nlp(text)
    clean_text = " ".join(token.text for token in doc if not token.is_stop)
    return clean_text

df['Spørsmål_uten_stopwords'] = df['Spørsmål'].apply(remove_stopwords)
df['Svar_uten_stopwords'] = df['Svar'].apply(remove_stopwords)


df.to_csv('renset_fil_1.csv', sep=';', index=False)



```

```python
from tensorflow.keras.layers import Embedding, Dense, GRU, InputLayer, TimeDistributed
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

embedding_dim = 128

model = Sequential([
    InputLayer(input_shape=(19,)),
    Embedding(total_words, embedding_dim, embeddings_initializer=tf.keras.initializers.random_normal(),
              trainable=True),  # Remove input_length=19
    GRU(256, return_sequences=True),
    TimeDistributed(Dense(33, activation='softmax'))
])

model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(),
    metrics=['accuracy']
)

model.summary()
```
