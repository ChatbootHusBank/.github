```python
import pandas as pd
data = pd.read_csv('Dataset_for_chatbot-1.csv', sep=';')
data.head()
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

# Assuming 'data' is your DataFrame with the "Spørsmål" column
# Replace 'data' with your actual DataFrame name

# Handling missing values by replacing NaN with an empty string
data["Spørsmål"].fillna("", inplace=True)

tokenizer = Tokenizer(oov_token='<oov>')
X = data["Spørsmål"]
tokenizer.fit_on_texts(X)
total_words = len(tokenizer.word_index) + 1

print(tokenizer.word_index)
print(total_words)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
sequences=tokenizer.texts_to_sequences(X)
# pad sequences 
max_sequence_len = max([len(x) for x in sequences])
input_sequences = np.array(pad_sequences(sequences,maxlen=30,padding='post'))
sequences
```

```python
input_sequences , input_sequences.shape
```

```python
#from tensorflow.keras.preprocessing.text import Tokenizer
data["Svar"].fillna("", inplace=True)
tokenizer2 = Tokenizer()
Y = data["Svar"]
tokenizer2.fit_on_texts(Y)
total_words2 = len(tokenizer2.word_index)+1
print(tokenizer2.word_index)
print(total_words2)
```

```python
sequences2=tokenizer2.texts_to_sequences(Y)
# pad sequences 
max_sequence_len = max([len(x) for x in sequences2])
output_sequences = np.array(pad_sequences(sequences2,maxlen=max_sequence_len,padding='post'))
output_sequences , output_sequences.shape
```

```python
# ONE_HOT ENCODING
import tensorflow as tf
ys = tf.keras.utils.to_categorical(output_sequences, num_classes=total_words2)
ys.shape 
```

```python
from tensorflow.keras.layers import Embedding, Dense, GRU, InputLayer, TimeDistributed
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

embedding_dim = 128

model = Sequential([
    InputLayer(input_shape=(19,)),
    Embedding(total_words, embedding_dim, embeddings_initializer=tf.keras.initializers.random_normal(),
              trainable=True),  # Remove input_length=19
    GRU(256, return_sequences=True),
    TimeDistributed(Dense(33, activation='softmax'))
])

model.compile(
    loss='categorical_crossentropy',
    optimizer=Adam(),
    metrics=['accuracy']
)

model.summary()
```
