```python
import pandas as pd
import spacy

nlp = spacy.load("nb_core_news_lg")

df = pd.read_csv('Dataset_for_chatbot_sisteverion-3.csv', delimiter=';')

# Funksjon for å fjerne stopwords
def remove_stopwords(text):
    if pd.isnull(text):
        return text
    doc = nlp(text)
    clean_text = " ".join(token.text for token in doc if not token.is_stop)
    return clean_text

df['Spørsmål_uten_stopwords'] = df['Spørsmål'].apply(remove_stopwords)
df['Svar_uten_stopwords'] = df['Svar'].apply(remove_stopwords)


df.to_csv('renset_fil_1.csv', sep=';', index=False)



```

```python
import re

cleaned_lines = []
with open('renset_fil_1.csv', 'r', encoding='utf-8') as file:
    for line in file:
        # Eksempel på å fjerne uønskede tegn, erstatt med faktisk logikk nødvendig for ditt datasett
        cleaned_line = re.sub(r'[^\x00-\x7F]+', '', line)  # Fjerner non-ASCII tegn som et eksempel
        cleaned_lines.append(cleaned_line)

# Skriv de rensede linjene tilbake til en ny fil
with open('cleaned_dataset.csv', 'w', encoding='utf-8') as file:
    file.writelines(cleaned_lines)
```

```python
import re

def clean_line_breaks_inside_quotes(text):
    # Dette regulære uttrykket matcher tekst innenfor anførselstegn
    pattern = r'"(.*?)"'
    # Funksjonen som skal brukes for å erstatte hver match
    def replace_line_breaks(match):
        # Erstatt alle linjeskift med et mellomrom
        return match.group(0).replace('\n', ' ')
    # Bruk re.sub() for å søke og erstatte alle tilfeller som matcher mønsteret
    return re.sub(pattern, replace_line_breaks, text, flags=re.DOTALL)

cleaned_lines = []
with open('renset_fil_1.csv', 'r', encoding='utf-8') as file:
    for line in file:
        cleaned_line = clean_line_breaks_inside_quotes(line)
        cleaned_lines.append(cleaned_line)

# Skriv de rensede linjene tilbake til en ny fil
with open('cleaned_dataset.csv', 'w', encoding='utf-8') as file:
    file.writelines(cleaned_lines)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from joblib import dump
import pandas as pd

# Load your dataset
data = pd.read_csv('cleaned_dataset.csv', delimiter=';', quotechar='"', engine='python', on_bad_lines='skip')

# Process your dataset
data['Spørsmål_uten_stopwords'].fillna('manglende', inplace=True)
data['combined_info'] = data[['Svar', 'Les mer her…', 'Link']].apply(lambda x: f"{x['Svar']} {x['Les mer her…']} {x['Link']}", axis=1)

X = data['Spørsmål_uten_stopwords']
y = data['combined_info']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)

# Initialize the tokenizer and model
tokenizer = AutoTokenizer.from_pretrained("sentence-transformers/roberta-large-nli-stsb-mean-tokens")
model = AutoModel.from_pretrained("sentence-transformers/roberta-large-nli-stsb-mean-tokens")

# Define the function to convert texts to embeddings
def get_embeddings(texts):
    inputs = tokenizer(texts.tolist(), return_tensors="pt", padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        # Ensure compatibility with different model outputs
        outputs = model(**inputs)
        # Averaging over the sequence length for each token to get the embedding
        embeddings = outputs.last_hidden_state.mean(dim=1).numpy()
    return embeddings

# Convert texts to embeddings
X_train_embeddings = get_embeddings(X_train)
X_test_embeddings = get_embeddings(X_test)

# Training and evaluation loop for the model
n_iterations = 2
for i in range(n_iterations):
    forest_model = RandomForestClassifier(random_state=40 + i)
    forest_model.fit(X_train_embeddings, y_train)
    y_pred = forest_model.predict(X_test_embeddings)
    accuracy = accuracy_score(y_test, y_pred)
    print(f'Accuracy of the model in iteration {i+1}: {accuracy}')

# Save the last model to a file after the loop
model_filename = 'final_model.joblib'
dump(forest_model, model_filename)
print(f'Final model saved to {model_filename}')
```
